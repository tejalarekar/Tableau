- A bar graph, also known as a bar chart, is a type of chart that uses bars to represent different categories of data and         the values associated with those categories. # 2. Line Graph

- A line graph, also known as a line chart, is a type of chart that uses a series of data points connected by straight lines to   show the changes in data over a period of time. The x-axis typically represents time, while the y-axis represents the data       values.

- Line graphs are often used to show trends in data over time, such as changes in stock prices, temperature, or population.They   are particularly useful when comparing multiple data sets over the same time period, as the lines on the graph can be easily     compared to see how they relate to each other.

- Line graphs are also useful to identify patterns, trends and seasonality in the data. They can also be used to forecast         future values using time series analysis techniques such as moving averages, exponential smoothing and ARIMA models.

- Line graphs can be created in various software tools like excel, R, Python and Tableau, etc.

# 3. Dual Axis Garph

- A dual-axis chart, also known as a combination chart or a multi-axis chart, is a type of chart that uses two or more axes to     display multiple types of data on the same chart. This allows for the comparison of different types of data in a single         visual, and can make it easier to spot patterns or trends that might not be as apparent in a single-axis chart.

- A typical dual-axis chart will have one axis on the left and one on the right side of the chart, but it can have more. Each     axis represents a different scale, and each data series is plotted on a different axis. The different scales allow for the       comparison of data series with different units or ranges.

# 4. Donut Chart

- A donut chart, also known as a doughnut chart, is a type of chart that is similar to a pie chart, but with a hole in the         center. The hole in the center makes it possible to display additional data or information in the center of the chart.
 
- Donut charts are often used to show the proportion of different categories within a whole. The size of each "slice" of the       donut corresponds to the proportion of the data it represents, and the different slices can be color-coded to make it easy to   distinguish between the categories.

- Donut charts can also be divided into multiple concentric circles, each showing a different level of detail. This can be         useful when you want to show hierarchical data or when you want to show multiple levels of data in the same chart.

- Donut charts are useful when you want to show a part-to-whole relationship and when you want to make a clear comparison         between the different categories. They are also useful when you want to show multiple levels of data in the same chart.         However, it can be challenging to compare the exact values of different sections, because the sections of a donut chart are     not easily quantifiable as in a Bar or a line chart.

# 5. Map/Symbol Map

- A map is a representation of geographic information, such as the location of countries, cities, landmarks, and other features   on the Earth's surface. Maps can be created in various forms, such as digital maps, paper maps, and online maps. They can also   be created in different styles, such as political maps, physical maps, and topographic maps.

- A symbol map is a type of map that uses symbols, such as icons or markers, to represent data points on the map. Symbol maps     are often used to display data that has a geographic component, such as the location of stores, the distribution of             population, or the location of natural resources.

# 6. Pie Chart

- A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In a pie         chart, the arc length of each slice (and consequently its central angle and area), is proportional to the quantity it           represents.

- Pie charts are often used to show the proportion of different categories within a whole. Each "slice" of the pie corresponds     to a different category, and the size of the slice is proportional to the percentage of the data that it represents. The         different slices can be color-coded to make it easy to distinguish between the categories.

- Pie charts are useful when you want to show a part-to-whole relationship and when you want to make a clear comparison between   the different categories. They are also useful when you want to show the proportion of different categories within a whole and   when you want to compare the exact values of different sections. However, it can be challenging to compare the exact values of   different sections when the number of slices are more than 5, in that case it's better to use a bar chart.

# 7. Word Cloud

- A word cloud, also known as a tag cloud or a text cloud, is a visual representation of the most frequently used words in a       given text or set of texts. The words are arranged in a way that the most frequently used words appear larger, and less         frequently used words appear smaller.

- Word clouds are often used to quickly identify the most common themes or topics within a text or set of texts. They can be       used to analyze social media posts, news articles, customer reviews, or any other type of text data. They are also commonly     used in text mining and natural language processing, to extract insights and patterns from large amounts of unstructured text   data.
